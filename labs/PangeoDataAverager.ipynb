```python
df_all = df.query("table_id == 'Amon' & variable_id == ['tas','pr'] ")
print(df_all.groupby('source_id')[["experiment_id", "variable_id", "table_id", "member_id"]].nunique())
```

```python
# find a set of ensemble members that all have the same experiments 

modelList=['MPI-ESM1-2-LR','IPSL-CM6A-LR','GISS-E2-1-G','CanESM5','EC-Earth3','CESM2']

expList = ['historical','ssp126','ssp370','ssp245','ssp585']

memberDict={}

for model in modelList:

    memberDict[model]=[]

    memberList=sorted(list(df_all[df_all.source_id==model].member_id.unique()), key=lambda x: (len(x),x))


    for member in memberList:
    
        hasExps=True

        df_temp = df_all[ np.array(df_all.source_id==model) & np.array(df_all.member_id==member) ]

        for exp in expList:
            if sum(df_temp.experiment_id == exp) == 0:
                hasExps=False
        
        if hasExps:
            memberDict[model].append(member)

            # cut it off after we have found this many members
            if len(memberDict[model])>=5:
                break

print(memberDict)
```

```python
# calculate global means - stolen from pangeo examples:
# https://gallery.pangeo.io/repos/pangeo-gallery/cmip6/global_mean_surface_temp.html

def get_lat_name(ds):
    for lat_name in ['lat', 'latitude']:
        if lat_name in ds.coords:
            return lat_name
    raise RuntimeError("Couldn't find a latitude coordinate")

def global_mean(ds):
    lat = ds[get_lat_name(ds)]
    weight = np.cos(np.deg2rad(lat))
    weight /= weight.mean()
    other_dims = set(ds.dims) - {'time'}
    return (ds * weight).mean(other_dims)
```

```python
varList=['tas','pr']

for model in modelList:
    for member in memberDict[model]:
        for var in varList:
            for exp in expList:
                
                print(model,member,var,exp)

                df_temp = df_all[ np.array(df_all.source_id==model) & \
                                  np.array(df_all.member_id==member) &\
                                  np.array(df_all.experiment_id==exp) & \
                                  np.array(df_all.variable_id==var) ]
                
                ds_temp = xr.open_zarr(fsspec.get_mapper(df_temp.zstore.values[0]), consolidated=True)

                outputName = f'output_timeseries/{var}_{exp}_{model}_{member}_glbAvg.nc'

                if os.path.isfile(outputName):
                    print(f'done {outputName} already')
                else:
                    global_mean(ds_temp).to_netcdf(outputName)
                    
```
